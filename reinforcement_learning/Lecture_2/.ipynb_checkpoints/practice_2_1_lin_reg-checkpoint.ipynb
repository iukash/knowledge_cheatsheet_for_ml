### Метод кросс-энтропии с глубоким обучением (способ с большей детализацией)

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(-5., 5., 100)
noise = torch.normal(0., 1., (1, 100))
y = 2*(x) + noise
plt.scatter(x, y.numpy())
plt.show()
```

```python
w = torch.tensor(0., requires_grad=True)
b = torch.tensor(0., requires_grad=True)
y_pred = b + x*w
plt.scatter(x, y.numpy());
plt.plot(x, y_pred.data, 'r');

loss = torch.mean((y - y_pred) ** 2)
print(loss.data, w.grad, b.grad)
```

```python
lr = 1e-1
loss.backward()
w.data = w.data - lr*w.grad
b.data = b.data - lr*b.grad
y_pred = b + x*w
loss = torch.mean((y - y_pred) ** 2)
print(loss)
w.grad.zero_()
b.grad.zero_()
plt.scatter(x, y.numpy());
plt.plot(x, y_pred.data, 'r');
```

```python
loss_list = []
lr = 1e-1

for _ in range(20):
    loss.backward()
    w.data = w.data - lr*w.grad
    b.data = b.data - lr*b.grad
    y_pred = b + x*w
    loss = torch.mean((y - y_pred) ** 2)
    loss_list.append(loss.data)
    w.grad.zero_()
    b.grad.zero_()
    
plt.scatter(x, y.numpy())
plt.plot(x, y_pred.data, 'r')
plt.show()
plt.plot(loss_list)
plt.title('Убывание функции потерь')
plt.show()
```
