### [Pandas](https://pandas.pydata.org/pandas-docs/stable/)  <a class="anchor" id="pandas"></a> `import pandas as pd`
* [agg()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.agg.html) - группировка с использованием одной или нескольких функций `groupby('name_gr').agg({'name_col': ['count', 'sum']}) `
* [apply()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html) - применяет функцию ко всем строкам/столбцам датафрейма `DataFrame.apply(func='')`
* [astype()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.astype.html) - изменить тип данных `convert_dict = {'col1': float, 'col2': str} DataFrame.astype(convert_dict)`
* [columns](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.columns.html) - возвращает или задает название столбцов датафрейма `DataFrame.columns`
* [count()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.count.html) - получить число строк/столбцов в датафрейме `DataFrame.count(axis=0)`
* [pd.concat()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.concat.html) - соединение датафреймов `pd.concat([df_1] + [df_2] * 4)`
* [drop_duplicates()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop_duplicates.html) - удаление повторяющихся строк `drop_duplicates()` inplace=False - удалять дубликаты на месте или возвращать копию
* [DataFrame](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html) - создать датафрейме `pd.DataFrame(data='данные', columns='название столбцов')`
* [DatetimeIndex](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DatetimeIndex.html) - объект для расчета по заданному периоду например month `pd.DatetimeIndex(data['col']).month`
* [dropna()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html) - удаляет пропущенные значения датафрейма `DataFrame.dropna(axis='columns')`
* [describe()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.describe.html) - вернуть таблицу из описательных статистик по датафрейму (чтобы получить статистики по категориальным столбцам необходимо выбрать только их) `DataFrame.select_dtypes(include=['object', 'datatime']).describe()`
* [duplicated()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.duplicated.html) - возвращает логический ряд, обозначающий повторящиеся строки `DataFrame.duplicated()#.sum()`
* [fillna()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html) - заполняет значения Na и NAN заданным значением или специальным методом `DataFrame.fillna(value=None, method=None)`
* [get_dummies](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html) - возвращает таблицу, закодированную по принципу One Hot Enconding OHE `pd.get_dummies(data, drop_first=False)`
* [groupby()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html) - группирует датафрейм по выбранной переменной `DataFrame.groupby(by='переменная').func(функция группировки)`
* [head()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.head.html) - вернуть первые n строк `DataFrame.head(n=5)`
* [info()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.info.html) - информация о датафрейме `DataFrame.info()`
* [insert()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.insert.html) - вставить столбец в DataFrame `DataFrame.insert(loc, column, value)`
* [isna()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.isna.html) - проверка датафрейма на пропущенные значения `DataFrame.isna()`
* [join()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.join.html) - объединение данных по индексу `DataFrame.join(other, on=None, how='left')` проверка размера - shape
* [loc[]](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html) - получить доступ к строкам/столбцам по условиям или значениям `DataFrame.loc[строка, столбец]`
* [mean()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.mean.html) - среднее значение по запрощенной оси `DataFrame.mean()`
* [median()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.median.html) - медианное значение по запрощенной оси `DataFrame.median()`
* [merge()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html) - соединение таблиц по общему столбцу `data.merge(subcat_dict, on='subcat_id', how='left')` проверка размера - shape
* [pivot_table()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html) - создать сводную таблицу `data.pivot_table(index=, columns='source', values=, aggfunc='sum')`
* [ProfileReport](https://pandas-profiling.github.io/pandas-profiling/docs/master/index.html) - много полезной информации о датафрейме (пропущенные значения, корреляции, уникальные значения, распределения и т.д.)
* [query()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.query.html) - метод сортировки `DataFrame.query('col1 == True or col2 > 20')`
* [read_csv()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html) - прочитать csv файл `pd.read_csv('путь и название файла', sep='разделитель', decimal='. или ,')`
* [read_excel()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_excel.html) - прочитать файл excel `pd.read_excel('файл.xlsx', sheet_name='первый лист')`
* [rename()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html) - переименовать столбец `DataFrame.rename(columns={"A": "a", "B": "c"})`
* [reset_index()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.reset_index.html) - сбросить значения индексов или уровней `reset_index()` drop=False - вставить столбец index или обновить текущий
* [shape](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.shape.html) - возвращает размер датафрейма `DataFrame.shape`
* [sample()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sample.html) - возвращает выборку заданного размера `DataFrame.sample(frac=1, replace=False, random_state=state)`
* [sum()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sum.html) - возвращает сумму значений по запрощенной оси датафрейма `DataFrame.sum()`
* [Series](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.html) - создать серию данных (одномерный массив) `pd.Series(data=''данные)`
* [sort_values()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sort_values.html) - сортировка значений (по умолчанию по возрастанию ascending) `DataFrame.sort_values(ascending=True)`
* [tail()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.tail.html) - вернуть последние n строк `DataFrame.tail(n=5)`
* [to_numeric()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_numeric.html) - преобразовывает аргумент в числовую переменную `pd.to_numeric(arg=, errors='raise', downcast=None)`
* [to_datetime()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_datetime.html) - преобразовать в объект дататайм `pd.to_datetime(arg, format='%d.%m.%YZ%H:%M:%S')` [format](https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior)

* [unique()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.unique.html) - возвращает уникальные элементы в порядке их появления `DataFrame.unique()`

* [value_counts()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.value_counts.html) - возвращает серию, содержащую количество уникальных значений `Series.value_counts(normalize=False)`

### [Scipy](https://docs.scipy.org/doc/) <a class="anchor" id="scipy"></a> `from scipy import stats as st`
* [norm()](https://docs.scipy.org/doc/scipy-0.18.1/reference/generated/scipy.stats.norm.html) - задать нормальное распределение с математическим ожиданием `mean` и СКО `std` `st.norm(mean, std)`
* [norm.ppf](https://practicum.yandex.ru/learn/data-scientist/courses/82786053-a3ac-47e6-b83b-b9226f67d284/sprints/21215/topics/378d0b2b-45f5-4e4b-901f-534e18d180b1/lessons/904687b6-0ec4-4760-af2e-93e72e523f23/) - вернет значение переменной для известной вероятности интервала слева от этого значения `norm.ppf(p=0.95)`
* [norm.cdf](https://practicum.yandex.ru/learn/data-scientist/courses/82786053-a3ac-47e6-b83b-b9226f67d284/sprints/21215/topics/378d0b2b-45f5-4e4b-901f-534e18d180b1/lessons/904687b6-0ec4-4760-af2e-93e72e523f23/) - вернет для известного значения вероятность интервала слева от этого значения `norm.cdf(x=1000)`
* [ttest_1samp()](https://practicum.yandex.ru/learn/data-scientist/courses/82786053-a3ac-47e6-b83b-b9226f67d284/sprints/21215/topics/0be764a0-2bfd-48b4-aecc-ffbfa98395c1/lessons/64fa4166-8b88-4bc6-87ec-31e58e599ed0/) - t-тест для проверки гипотезы равенста среднего заданному числу `st.ttest_1samp(array, value)` в случае проверки односторонней гипотезы `rezult.pvalue / 2`
* [ttest_ind()](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html)- t-тест для проверки гипотезы равенста средних 2-х независимых выборок генеральных совокупностей `scipy.stats.ttest_ind(sample_1, sample_2, equal_var=True)` `equal_var` - параметр о том считать ли равные дисперсии в выборках: генеральные совокупности не должны зависеть друг от друга и выборочные средние должны быть нормально распределены
* [ttest_rel()](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_rel.html)- t-тест для проверки гипотезы равенста средних 2-х зависимых выборок ген. совокупностей `st.ttest_rel(before, after)`
* [norm.interval](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.norm.html#scipy.stats.norm) - доверительный интервал нормального распределения `stats.norm.interval(0.68, loc=mean, scale=sigma)`
* [stats.t.interval()](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html) - доверительный интервал t-распределения `stats.t.interval(alpha, df, loc=0, scale=1)`

### <a class="anchor" id="alg"></a> Algoritms
* [сортировка подсчетом и словари](https://www.youtube.com/watch?v=Nb5mW1yWVSs&t=3227s) - сортировка подсчетом заключается в создании массива нулей уникальных значений и проходом по массиву значение нужного элемента массива увеличивать на 1 (на малое количество целесообразно использовать именно сортировку, на большое уже словарь)
* [префиксные суммы](https://www.youtube.com/watch?v=de28y8Dcvkg&t=3265s) - алгоритм решает задачу по нахождению суммы на интревале путем последовательного вычисления сумм элементов (и сохранении этого ряда) и вычисления суммы путем вычитания из правой границы левой (используя построенный ряд и заданные индексы) 
* [два указателя](https://www.youtube.com/watch?v=de28y8Dcvkg&t=3265s) - алгоритм основан на прохождении по двум отсортированным массивам путем движения итератора (индекса) левого с одновременным движением правого (по условию в зависимости от сортировки)
```
def func_two_uk(sortednums, k):
    cnt = last = 0
    for first in range(len(sortednums)):
        while last < len(sortednums) and sortednums[last] - sortednums[first] <= k
            last += 1
        cnt += len(sortednums) - last
    return cnt
```
* [бинарный поиск](https://www.youtube.com/watch?v=YENpZexHfuk&t=2026s) - поиск элемента из предположения, что есть событие которое на отрезке дает отрицательный результат, а на остальной длине положительный и задача заключается в нахождении элемента перехода, тогда задачу можно решать последовательным отбрасыванием половины массива т.е. смотрим значение в середине и в зависимости от искомого значения смещаем либо левую либо правую границу в середину и так далее уменьшая вполовину до нахождения. сложность алгоритма log(N).
